```python
from pyspark.sql import SparkSession

# 1. Crear una sesión de Spark apuntando al Master de Docker Compose
# El Master se llama 'spark-master' y usa el puerto 7077
spark = SparkSession.builder \
    .appName("PySpark-Test-Connection") \
    .master("spark://spark-master:7077") \
    .getOrCreate()

print("✅ SparkSession iniciada correctamente. Conectado al Master.")
```
