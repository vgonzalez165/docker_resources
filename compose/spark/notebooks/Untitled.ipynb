```python
from pyspark.sql import SparkSession

# 1. Crear una sesión de Spark apuntando al Master de Docker Compose
# El Master se llama 'spark-master' y usa el puerto 7077
spark = SparkSession.builder \
    .appName("PySpark-Test-Connection") \
    .master("spark://spark-master:7077") \
    .getOrCreate()

print("✅ SparkSession iniciada correctamente. Conectado al Master.")
```

```python
# 2. Crear un DataFrame simple para testear la funcionalidad
test_data = [
    ("Alice", 1), 
    ("Bob", 2), 
    ("Charlie", 3), 
    ("David", 4)
]
columns = ["name", "id"]

df = spark.createDataFrame(test_data, columns)

# 3. Lanzar una acción Spark (esto fuerza la ejecución en los Workers)
print("------------------------------------------------------------------")
print("Total de filas en el DataFrame de prueba (debe ser 4):")
print(df.count())

print("\nVista previa del DataFrame:")
df.show()

# 4. Detener la sesión
spark.stop()
print("------------------------------------------------------------------")
print("✅ Test completado. La conexión y la ejecución distribuida funcionan.")
```
