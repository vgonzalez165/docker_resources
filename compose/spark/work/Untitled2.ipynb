```python
from pyspark.sql import SparkSession

# 1. Crear una sesi√≥n de Spark apuntando al Master
# Usa 'spark-master' como hostname (el nombre del servicio en Docker Compose)
SPARK_MASTER_URL = "spark://spark-master:7077"

try:
    spark = ( SparkSession.builder 
                .appName("PySpark-Conexion-Test") 
                .master(SPARK_MASTER_URL) 
                .config("spark.ui.port", "4040") 
                .getOrCreate() 
            )
    
    print(f"‚úÖ SparkSession iniciada correctamente. Conectado a: {SPARK_MASTER_URL}")

    # 2. Test Funcional: Crea un DataFrame y fuerza una acci√≥n distribuida (count)
    # Si esta l√≠nea funciona, los Workers est√°n recibiendo y procesando tareas.
    test_data = [
        ("Vuelo A", 100), 
        ("Vuelo B", 200), 
        ("Vuelo C", 300), 
        ("Vuelo D", 400)
    ]
    columns = ["vuelo", "pasajeros"]
    df = spark.createDataFrame(test_data, columns)

    total_filas = df.count()

    print("--------------------------------------------------")
    print(f"Total de filas procesadas (Prueba de Worker): {total_filas}")
    print("\nVista previa del DataFrame:")
    df.show()
    print("--------------------------------------------------")
    print("üöÄ La conexi√≥n y la ejecuci√≥n distribuida son exitosas.")

except Exception as e:
    print(f"‚ùå ERROR: Fall√≥ la conexi√≥n o la ejecuci√≥n de Spark.")
    print(e)

finally:
    # 3. Detener la sesi√≥n para liberar recursos
    if 'spark' in locals() and spark.sparkContext._jsc is not None:
        spark.stop()
```
